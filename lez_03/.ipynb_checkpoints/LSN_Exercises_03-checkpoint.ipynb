### Exercise 3.1 <span style="color:red"> Plain vanilla option pricing </span>
Black-Scholes theory assumes that the time evolution of the asset price is stochastic and exhibits geometric Brownian motion (GBM) with constant risk-free interest rate $r$ and volatility $\sigma$.

By using the following parameters: 
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$

compute at time $t=0$ **via Monte Carlo** the European call-option price, $C[S(0),0]$, and put-option price, $P[S(0),0]$:

1. <span style="color:red"> By sampling **directly** the final asset price $S(T)$</span> for a $GBM(r,\sigma^2)$
2. <span style="color:red"> By sampling the discretized $GBM(r,\sigma^2)$ path of the asset price</span> 
dividing $[0,T]$ in $10^2$ time intervals: $S(0)$, $S(t_1)$, $S(t_2)$, ... up to $S(t_{100}=T=1)$

### For comparison, Black-Scholes analytic solution:

```python
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

# Results

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

N=100 #block number
x = np.arange(N)
throw_no = [4,5,6]
ev_type=["direct","discrete"]
value_name=["call","put"]
value_type=[call,put]
color_cycle=[['xkcd:forest green','xkcd:grass green','xkcd:spring green'],['xkcd:crimson', 'xkcd:orange','xkcd:bright yellow']]
```

The previous code, kindly provided by prof. Galli, estimates exactly the optimal value for a call and a put offer for a starting price of $S(0) = 100$ at time $T=1$. The following graphs depict the same value obtained with numerical simulations of a GMB with parameters $r=0.1$ risk interest rate and $\sigma^2=0.25$ volatility

```python
fig=plt.figure(figsize=[8,6])
colour_ = ['xkcd:sky blue', 'xkcd:golden yellow']

for i in [0,1]:
    file_name = "data_3_1/"+ev_type[i] + "_distribution.txt"
    distr=np.loadtxt(file_name)
    plt.hist(distr, bins=100, color = colour_[i], alpha=0.5, label=ev_type[i])
    
plt.legend(fontsize=14)
plt.title("Distribution of S(T)", fontsize=16)
plt.xlabel('price')
plt.ylabel('frequency of price in an interval $T$')
```

The first to notice is that the general distribution of prices $S(T)$ is obviously not homogeneous, but neither symmetric: the price tends to grow after a time T. Due to this asymmmetry, we expect the final price to be more often greater than the strike price. The call option evaluation $C= max[0,S(T)-K]$ tends to take non-trivial values more frequently that the put option. However, due to the large number of the total events, the statistics for both is more than large enough to obtain relevant results.

Secondly, the distribution of the final value is practically the same for both methods of estimating $S(T)$: we expect that the estimation of the optimal pricing option does not depend on the method of simulation.

```python
# figure plot of different combinations of data
fig,axs = plt.subplots(2,2, figsize=[20,15], sharex=True, sharey='row')
fig.suptitle('Progressive average of respective optimal price', fontsize=16)

for j in [0,1]:
    for i in [0,1]:
        for n in throw_no:
            file_name = "data_3_1/" + ev_type[j] + "_" + value_name[i] + "_" + str(n) +".txt"
            data, err = np.loadtxt(file_name, usecols=(0,1), unpack='true')
            label_n = "10^"+ str(n) + " total throws"
            axs[i][j].errorbar(x, data, yerr=err, fmt='o',  markersize=2, label = label_n,color=color_cycle[j][n-4])
        
        label_line = "Exact value of "+ value_name[i]
        axs[i][j].axhline(y=value_type[i], color='xkcd:black', label = label_line)
        axs[i][j].set_xlabel('# block')
        axs[i][j].set_ylabel(value_name[i])
        title_n = ev_type[j] + " " + value_name[i]
        axs[i][j].set_title(title_n, fontsize=14)
        axs[i][j].grid(True)
        axs[i][j].legend(fontsize='large')
```

The previous graphs show the progressive simulation outcome for optimal call or put offered compared to the exact value. The green graphs show the results obtained by computing the final price directly, while the red results are obtained by simulating a stochastic process of 100 uniform time steps between $T_0 = 0$ and $T_1=1$. 
As we can see, they are all statistically compatible with the expected value. By increasing the number of total throws the estimates and uncertainties narrow around the expected value; $10^4$ throws are sufficient to reach obtain acceptable estimates. 

```python
file_name = ["data_3_1/discrete_call_5.txt","data_3_1/discrete_call_1.txt","data_3_1/discrete_call_2.txt"]
labels = ["100 steps", "200 steps", "1000 steps"]
i=0

fig,axs = plt.subplots(1,2, figsize=[18,6],gridspec_kw={'width_ratios': [2, 1]})

for name in file_name:
    data, err = np.loadtxt(name, usecols=(0,1), unpack='true')
    axs[0].errorbar(x, data, yerr=err, fmt='o',  markersize=2, label = labels[i])
    i+=1
    
axs[0].set_xlabel('# block')
axs[0].set_ylabel("call value")
axs[0].axhline(y=call, color='xkcd:black', label = "true value of call")

axs[0].set_title("Comparison of call value for different number of intermediate steps", fontsize=14)
axs[0].grid(True)
axs[0].legend(fontsize='large')

# ditribution of price value 
file_name = ["data_3_1/discrete_distribution.txt","data_3_1/discrete_distribution_1.txt","data_3_1/discrete_distribution_2.txt"]
colour_ = ['xkcd:pumpkin','xkcd:golden yellow', 'xkcd:yellowish']
i=0

for name in file_name:
    distr=np.loadtxt(name)
    axs[1].hist(distr, bins=100, color = colour_[i], alpha=0.5, label=labels[i])
    i+=1
    
axs[1].legend(fontsize=14)
axs[1].set_title("Distribution of S(T)", fontsize=16)
plt.show()
```

The discrete steps seemingly reach the expected value faster in this simulation; however, one can see that actually the results are statistically compatible . For this "plain vanilla" option pricing there is no optimal number of steps to divide the GBM into: each simulation in the previous graph uses a different number of steps for $10^5$ different walks, but the average values are clearly equivalent. What changes is the sample size of calls, but the distribution remains the same (as shown in the second graph).
This is because the stochastic method used to predict the following value is exact for any time step, be it the whole interval or a minimal fraction. The possibility of considering intermediate steps is however more relevant when using more "exotic" pricing opions, which usually also depend on time.
