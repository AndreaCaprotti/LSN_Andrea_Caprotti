### Exercise 2.1
- Compute the following 1D integral via Monte Carlo 
    $$I = \int_0^1 \frac{\pi}{2} \cos ( \dfrac{\pi x}{2}) dx = 1$$
    1. sampling a uniform distribution in $[0,1]$
    2. using importance sampling (i.e. sampling a non-uniform probability in $[0,1]$)

An integral can be estimated numerically following Riemann's definition $$I = \int_a^b f(x) d x = \lim_{N \to \infty} \sum_{i=0}^N f(x_i) = (b-a) \langle f \rangle _{[a,b]}$$ 
The mean can be estimated from random variables ${x_i}$ uniformly distributed in the interval $[a,b]$. Using Monte Carlo methods to generate the random variables, different estimates of the integral are obtained. The final estimate and statistical uncertainty are determined with a block average of the sample. 

Given an integral in the form $ \int_a^b g(x) p(x) d x$, where $p(x)$ is a probability distribution, the numerical evaluation can be simplified as the average of $g(x)$ with the random values $x = \{ x_0 \dots x_N \}$ sampled following $p(x)$. However, this distribution is not always the best choice, as perhaps it is hard to sample computationally -for example when its cumulant is unknown-. One can still sample a generic $f(x)$ distribution and compute the integral as the average of $\frac{g(x)p(x)}{f(x)}$. The better $f(x)$ approximates the integrand in the given interval, the faster the numerical evaluation converges to the actual integral value: 
$$ G_N = \frac{1}{N}\sum_{i=1}^N \frac{g(x)p(x)}{f(x)} = \frac{1}{N}\sum_{i=1}^N \frac{g(x)p(x)}{g(x)p(x)}I = I$$ 

This method, known as *importance sampling*, helps speeding up the numerical computation of integrals.
The following example shows the effective improvement in the estimation thanks to a better sampling choice.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
from scipy import optimize
```

```python
ave,err = np.loadtxt("02_1/std_integral.txt",usecols=(0,1), unpack='true')
ave_2,err_2 = np.loadtxt("02_1/sample_integral.txt",usecols=(0,1), unpack='true')
x = np.arange(100)

fig,axs = plt.subplots(1,2, figsize=[16,6],gridspec_kw={'width_ratios': [2, 1]})

axs[0].axhline(y=1, color='r', label = 'real value')
axs[0].errorbar(x,ave,yerr=err, label='uniform sample')
axs[0].errorbar(x,ave_2,yerr=err_2, label='importance sample')

axs[0].legend(loc = 'lower right')
axs[0].set_xlabel('# blocks')
axs[0].set_ylabel('$<r> $')
axs[0].set_title('Comparison of progressive estimate of definite integral using two different sampling methods')
axs[0].grid(True)

# integrand function and its approximations
x = np.linspace(0.0,1.0,num=100)
f_x = np.pi * np.cos(np.pi/2 * x)/2
taylor = np.pi*(1 - x**2 / 2)/2
lin = 2*(1-x)

axs[1].plot(x, f_x, label='integrand')
axs[1].plot(x,taylor, label = 'Taylor expansion to the first order')
axs[1].plot(x, lin, label = 'linear approximation')
axs[1].set_title('Integrand and approximations')
axs[1].set_xlabel('x')
axs[1].set_xlim([0,1])
axs[1].set_ylim([0,2])
axs[1].set_ylabel('y')
axs[1].legend(loc = 'lower left')
axs[1].grid()

plt.show()
```

The previous graps shows the estimate of the definite integral $$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$ computed with uniform sampling and importance sampling.

For the importance sampling, I first tried to find the best estimate to approximate the integrand. Since sampling $\frac{\pi}{2}\cos(\pi x/2)$ would be quite hard, I first tried to expand the function as the Taylor series around $x=1$ resulting in the function $\frac{\pi}{2} ( 1 - \frac{1}{2} x^2 )$. However, this function cannot be used as a probability distribution: its cumulative is not invertible on the needed interval. Therefore, I chose a simpler linear approximation $y=2(1-x)$, which has a known cumulant that can be easely inverted. This linear distribution can be sampled as a function $ x = 1-\sqrt{1-y}$ of uniformely distibuted random variables $y \in [0,1]$. Using these random values $x$ the integral is determined as the the average of 
$$ \frac{\pi}{2} \frac{2 \cos (\pi / 2 x)}{1-x}$$

It is easely seen that the importance sampling converges much faster and with smaller uncertainty to the actual value of the integral than the basic uniform sampling case: the estimate is compatible after just around 50 blocks, while the uniform sampling with 100 blocks is compatible with a $2\sigma$ discrepancy. 

In order to increase the speed of convergence, one would have to find a probability distribution that better approximates the integrand. However, this is easier said than done: any easy polynomial with order grater than 1 has the problem of a non-invertible cumulant on the interval $[0,1]$, where the variables need to be sampled in order to compute the integral. I dare not to venture in trying to sample trigonometric or hyperbolic approximations of the function, therefore we will appreciate the evident speed-up with this "plain vanilla" example of importance sampling.

### Exercise 2.2

- 3D Random Walks (RW) on a cubic lattice and in the continuum: Repeat many times (e.g. say $10^4$) the simulation of a random walk in 3D always starting at the origin
    1. on a cubic lattice with lattice constant $a=1$; at each discrete time the walker makes a forward or backward step of length equal to $a$ in one of the 3 principal directions of the lattice: $x$, $y$ or $z$
    2. in the continuum; at each discrete time the walker makes a step of length equal to $a=1$ along a **random direction** obtained by sampling **uniformly** the solid angle: $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$

<span style="color:red">Show a picture of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ for both RWs</span>, with their statistical uncertainties, as function of the step $i\in [0,10^2]$.


```python
N=100
x=np.arange(N+1)

def test_func (x, a):
    return a * (x**(1/2))

# load walk examples
d_1 = np.loadtxt("data_2_2/walk_example_d.txt",usecols=(0,1,2),unpack='true')
d_2 = np.loadtxt("data_2_2/walk_example_d_1.txt",usecols=(0,1,2),unpack='true')
d_3 = np.loadtxt("data_2_2/walk_example_d_2.txt",usecols=(0,1,2),unpack='true')

c_1 = np.loadtxt("data_2_2/walk_example_c.txt",usecols=(0,1,2),unpack='true')
c_2 = np.loadtxt("data_2_2/walk_example_c_1.txt",usecols=(0,1,2),unpack='true')
c_3 = np.loadtxt("data_2_2/walk_example_c_2.txt",usecols=(0,1,2),unpack='true')
```

```python
fig = plt.figure(figsize=[20,8])
fig.suptitle('Comparison of generic random walks', fontsize = 16)
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.set_title ('discrete walks')
ax.plot3D(*d_1, color = 'xkcd:golden yellow')
ax.plot3D(*d_2, color = 'xkcd:bright red')
ax.plot3D(*d_3, color = 'xkcd:burnt orange')
ax.view_init(5, 20)

ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.set_title ('continuous walks')
ax.plot3D(*c_1, color='xkcd:azure')
ax.plot3D(*c_2, color='xkcd:kelly green')
ax.plot3D(*c_3, color='xkcd:bright blue')
ax.view_init(5, 20)

plt.show()
```

These first two graphs show the trajecotries of two generic random walk: the ones on the left progresse with discrete steps of length $a=1$ on one of the three reference axes chosen at random; the ones on the right instead proceed with steps of fixed length $a=1$ but in any spatial direction. In these examples the discrete walks may seem more dispersive than their continuous counterparts. However, this is just an impression due to the more ordered appearence of the discrete walks. It can  be noticed right away from the graphs that the red walk on the right moves very little around the origin, while the blue graph on the right spans a much larger distance. 
It will be shown shortly afterwards, on average the mean distance of the two types of walks grows like $\sqrt{t}$, where $t$ represents the number of steps. 

```python
# discrete walk analysis
ave_d, err_d = np.loadtxt("data_2_2/discrete_walk.txt",usecols=(0,1),unpack='true')
ave_d_nb, err_d_nb = np.loadtxt("data_2_2/discrete_walk_nb.txt",usecols=(0,1),unpack='true')
params = optimize.curve_fit(test_func, x, ave_d)
fig = plt.figure(figsize=[18,7])
plt.errorbar(x, ave_d_nb, yerr=err_d_nb, fmt='o',label='average on $10^4$ walks',c='xkcd:sky blue')

plt.plot(x, test_func(x, params[0]), label='Fitted function', c='k')
for i in range(7):
    b_of_w = np.loadtxt("data_2_2/bunch_of_walks.txt", usecols=(i), unpack = 'true')
    x = np.arange(len(b_of_w))
    plt.scatter(x, b_of_w, label = "randomly chosen walk no." +str(i+1))

x=np.arange(N+1)
#plt.errorbar(x, ave_d, yerr=err_d, fmt='x',label='discrete steps, block average')

plt.legend()
plt.xlabel('# steps')
plt.ylabel('$\sqrt{<r^2>}$')
plt.title('Distance from origin for different discrete walks ')
plt.grid(True)
```

This graph compares the distance from the origin of different discrete random walks after each step. 
The values are then compared with the average of $10^4$ different walks, which follows quite well the fit with a square root function in the number of steps. The errorbars represent the total variance (estimated on the whole sample, **not** using the blocking method) in order to show the average dispersion of the walks. We can appreciate quite well the variety of different walks, as some tend to remain around the origin much longer while others move much further than the average; the majority tends to remain confortably inside the standard deviation, as we would expect. 

```python
fig, axs = plt.subplots(1,2,figsize=[18,7])
fig.suptitle('Comparison of fit with averge distance from origin')

params = optimize.curve_fit(test_func, x, ave_d)
axs[0].plot(x, test_func(x, params[0]), label='Fitted function for discrete walk')
axs[0].errorbar(x, ave_d, yerr=err_d, label='discrete walk, block average')
axs[0].legend()
axs[0].set_xlabel('# steps')
axs[0].set_ylabel('$\sqrt{<r^2>}$')
axs[0].set_title('Discrete walks')
axs[0].grid(True)

ave_c, err_c = np.loadtxt("data_2_2/continuous_walk.txt",usecols=(0,1),unpack='true')
ave_c_nb, err_c_nb = np.loadtxt("data_2_2/continuous_walk_nb.txt",usecols=(0,1),unpack='true')
params = optimize.curve_fit(test_func, x, ave_c)

axs[1].plot(x, test_func(x, params[0]), label='Fitted function for continuous walks', c = 'xkcd:kelly green')
axs[1].errorbar(x, ave_c, yerr=err_c, label='continuous walk, block average', c = 'xkcd:bright red')
axs[1].legend()
axs[1].set_xlabel('# steps')
axs[1].set_ylabel('$\sqrt{<r^2>}$')
axs[1].set_title('Continuous walks')
axs[1].grid(True)
```

These final graphs compare the average distance from the origin for both kind of walks . As we can see, both for discrete steps and for continuous steps the agreement with their fitted square root functions is almost perfect, even considering the very small uncertainty (this time determined by dividing the $10^4$ walks in 100 blocks).
