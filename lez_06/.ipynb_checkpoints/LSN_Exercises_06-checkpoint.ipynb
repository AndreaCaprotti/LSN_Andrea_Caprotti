In Statistical Mechanics the one-dimensional (1D) Ising Model is exactly resoluble. The Hamiltonian for a system of N spins with nearest neighbors interaction is given by (in the following I will assume units such that $\mu_B=1$ and $k_B=1$):

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N (s_i + s_{i+1}) 
$$

where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model),and with the assumption $s_{N+1} = s_{1}$.

The partition function for $N$ spins, $Z$, can be exactly computed and it turns out to be:

$$
Z=\sum_{\{ s_i \}} \exp \left( -\beta H\right) = \lambda_1^N + \lambda_2^N
$$

where
$\lambda_{1,2} = 
e^{\beta J} 
\cosh (\beta h) \pm \left[ e^{2\beta J} \cosh^2 (\beta h) - 2\sinh (2\beta J) \right]^{1/2}$, 
and where $\beta=\frac{1}{k_b T} = \frac{1}{T}$.

```python
import numpy as np
import matplotlib.pyplot as plt
points=100
T = np.linspace(0.4,2.1,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
```

### Exercise 06.1

Simulate 1D Ising model with 
- the Metropolis sampling algorithm and
- the Gibbs sampling algorithm

to sample the Boltzmann's weight $p(\{ s_i \}) = \exp(-\beta H)/Z$,
adding also the possibility to restart from a previous spin configuration.

Compute:
1. $U(N,T) = \langle H \rangle_{h=0}$
2. $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 
3. $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$
4. $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$

as a function of $T$ in the range $\left[0.5;2.0\right]$ for a system of $N=50$ spins with $J=1$.

The first configuration is chosen at random with 50-50 chance for both up and down spins; the temperature influences solely how the moves are accepted, both for Metropolis and for Gibbs acceptance. The results for different temperatures will be compared to the exact expected values of the different quantities, which in this case can computed analytically since the partition function is known exactly.

The simulation itself ran over 200 blocks of 5000 Monte Carlo steps each, which consisted in the attempt of 50 random moves (ideally, one per each spin site).

```python
fig, axs = plt.subplots(2, 1, figsize=[20,12], sharex=True)

#Metropolis, T=0.5
U = np.loadtxt("ISING_data/output.energy_0.500000_Metropolis.txt", usecols=(1), unpack='true')
n_block = np.arange(len(U))
axs[0].plot(n_block, U, label = "Metropolis evolution at $T^{\star} =0.5$")
avg_en = sum(U) / len(U)
axs[1].plot(n_block, U - avg_en, label = "Metropolis evolution at $T^{\star} =0.5$")

U = np.loadtxt("ISING_data/output.energy_0.725000_Gibbs.txt", usecols=(1), unpack='true')
axs[0].plot(n_block, U, label = "Gibbs evolution at $T^{\star} =0.725$")
avg_en = sum(U) / len(U)
#axs[1].plot(n_block, U - avg_en, label = "Metropolis evolution at $T^{\star} =0.5$")
U = np.loadtxt("ISING_data/output.energy_0.875000_Metropolis.txt", usecols=(1), unpack='true')
axs[0].plot(n_block, U, label = "Gibbs evolution at $T^{\star} =0.725$")
avg_en = sum(U) / len(U)
#axs[1].plot(n_block, U - avg_en, label = "Metropolis evolution at $T^{\star} =0.5$")

#Gibbs, T=1.025
U = np.loadtxt("ISING_data/output.energy_1.025000_Gibbs.txt", usecols=(1), unpack='true')
n_block = np.arange(len(U))
avg_en = sum(U) / len(U)
axs[0].plot(n_block, U, label = "Gibbs evolution at $T^{\star} =1.025$")
axs[1].plot(n_block, U - avg_en, label = "Gibbs evolution at $T^{\star} =1.025$")

axs[0].set_title('Ising 1D, Instantaneous energy')
axs[1].set_title('Ising 1D, Difference between instantaneous energy and average energy')
plt.xlabel('no of step')
axs[0].set_ylabel('$U$')
axs[1].set_ylabel('$U - <U>$')
axs[0].grid(True)
axs[1].grid(True)
axs[0].legend()
axs[1].legend()
plt.show()
```

The first thing to consider is the equilibration phase: in order to execute properly the Metropolis algorithm, the system needs to be at constant temperature, that is at the equilibrium. I chose to look at the internal energy as an indicator of the thermalisation, since temperature is not directly accessible. When the equilibrium is reached, the instantaneous energy should be compatible with the average energy, whilst during the transient one would expect greater fluctuations due to the fact that the configuration is constantly changing. 
The thermalisation was observed by having short simulations, of a couple of hundreds blocks of one step each, in which the energy was monitored continuously. 

The top graph shows clearly how the internal energy at different temperatures remains almost constant from the beginning up until the end. The bottom graph shows instead the difference between the instantaneous energy and the average of the energy during the whole run. This corresponds roughly to the square root of the variance, in order to show instantaneous deviation from the mean. We can see that the deviation fluctuates around 0 right from the start. 

The system appears to thermalise quite quickly at any temperature, therefore the equilibration phase does not need to take too long. However, just to be sure, the following data was all obtained after an equilibration phase of about $N \approx 10^5$ steps (corresponding to a full simulation without measurements)

The simulations were run using two different kinds of acceptance: 
- the Metropolis algorithm chooses randomly a new configuration and decides to accept it depending on the Boltzmann probability of the new configuration. Given a uniform transition probability, if the new configuratoin is more probable than the previous the step is always accepted, otherwise a random test is done with the ratio of the new probability with the previous: $x_k < \frac{p_{New}}{p_{Old}}$
- the Gibbs algorithm, on the other hand, can be seen as a Metropolis algorithm that always accepts the move. At every step it picks a random position, determines the Boltzmann probability for that position both for spin up and spin down and chooses which value to keep based on the conditional probability: 
$$ p(s_k = \pm 1 | \{ s_j , j\neq k \} ) = \dfrac{p_{joint}}{p_{marginal}} = \dfrac{1}{1 + e^{\mp 2 \beta J \Sigma_i s_i}} $$
    Since there are only two possible alternatives, a random test is done with $ p ( s_k = 1 | \{ s_j , j\neq k \})$: if it passes, $s_k = 1$, otherwise $s_k = -1$. The main difference with the Metropolis is that the acceptance does not depend on the previous step but only on the current configuration.

```python
fig, axs = plt.subplots(2,2, figsize = [20,15])

#energy
e = -J*( th + ch*thN )/( 1 + thN )     #exact solution for energy dependency on temperature  
axs[0][0].plot(T, e, label = 'exact solution')

temp, U_met , U_err_met = np.loadtxt("ISING_data/temp_compare_Metropolis.txt", usecols=(0,1,2), unpack='true')
axs[0][0].errorbar(temp, U_met, yerr=U_err_met, fmt='o', label ='Metropolis simulation')

temp, U_gib , U_err_gib = np.loadtxt("ISING_data/temp_compare_Gibbs.txt", usecols=(0,1,2), unpack='true')
axs[0][0].errorbar(temp, U_gib, yerr=U_err_gib, fmt='x', label ='Gibbs simulation')

axs[0][0].legend()
axs[0][0].set_title('Ising 1D, internal energy')

axs[0][0].set_xlabel('$T^{\star}$')
axs[0][0].set_ylabel('$U/N$')
axs[0][0].grid(True)

#heat capacity
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
axs[0][1].plot(T, heat, label ='exact solution')

temp, c_met , c_err_met = np.loadtxt("ISING_data/temp_compare_Metropolis.txt", usecols=(0,3,4), unpack='true')
axs[0][1].errorbar(temp, c_met, yerr=c_err_met, fmt='o',label ='Metropolis simulation')

temp, c_gib , c_err_gib = np.loadtxt("ISING_data/temp_compare_Gibbs.txt", usecols=(0,3,4), unpack='true')
axs[0][1].errorbar(temp, c_gib, yerr=c_err_gib, fmt='x', label ='Gibbs simulation')

axs[0][1].legend()
axs[0][1].set_title('Ising 1D, Heat capacity')
axs[0][1].set_xlabel('$T^{\star}$')
axs[0][1].set_ylabel('$C$')
axs[0][1].grid(True)

#magnetisation (h=0.02)
h=0.02 #external field
b = 1/T
    
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)
axs[1][0].plot(T, M, label='exact solution')

temp, m_met , m_err_met = np.loadtxt("ISING_data/temp_mag_Metropolis.txt", usecols=(0,1,2), unpack='true')
axs[1][0].errorbar(temp, m_met, yerr=m_err_met, fmt='o',label ='Metropolis moves')

temp, m_gib , m_err_gib = np.loadtxt("ISING_data/temp_mag_Gibbs.txt", usecols=(0,1,2), unpack='true')
axs[1][0].errorbar(temp, m_gib, yerr=m_err_gib, fmt='x', label ='Gibbs moves')
axs[1][0].legend()
axs[1][0].set_title('Ising 1D, magnetization M with h = 0.02')
axs[1][0].set_xlabel('$T^{\star}$')
axs[1][0].set_ylabel('$M$')
axs[1][0].grid(True)

# susceptibility
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
axs[1][1].plot(T, X, label='exact solution')

temp, X_met , X_err_met = np.loadtxt("ISING_data/temp_compare_Metropolis.txt", usecols=(0,5,6), unpack='true')
axs[1][1].errorbar(temp, X_met, yerr=X_err_met, fmt='o', label ='Metropolis moves')
temp, X_gib , X_err_gib = np.loadtxt("ISING_data/temp_compare_Gibbs.txt", usecols=(0,5,6), unpack='true')
axs[1][1].errorbar(temp, X_met, yerr=X_err_met, fmt='x', label ='Gibbs moves')

axs[1][1].set_ylim([0,100])
axs[1][1].legend()
axs[1][1].set_title('Ising 1D, Magnetic susceptibility')
axs[1][1].set_xlabel('$T^{\star}$')
axs[1][1].set_ylabel('$\chi$')
axs[1][1].grid(True)
plt.show()
```

The graphs show the results for internal energy and magnetisation per spin, heat capacity and magnetic susceptibility. Energy is determined at each step using the Ising hamiltonian, while the magnetisation is simply the sum of all spins. The heat capacity is the specific heat per spin; the latter is estimated as the variace of the total internal energy and only afterwards it is divided by the total number of spins (a detail that may have caused some issues at the beginnig). The same observation needs to be made also for the magnetic susceptibility, which corresponds to the fluctuations of the total magnetisation.

Both Metropolis and Gibbs algorithms present very low uncertainties and quite a good agreement with the theoretical value, determined from the partition function. At first, there was a slight difference between the expected values and the measured data, especially at low temperatures. This is mainly due to the very low acceptance of the algorithms: the system tends to remain in a very stable configuration which is hard to perturb. Any move that might induce some sort of instability is rarely accepted, when it does it generates high fluctuations in the estimation of the relevant quantity.

As we can see in the previous graphs, the acceptance rate grows constantly with temperature. The Boltzmann distribution, used for acceptance of the moves, heavily penalises lower temperatures, while it tends to accept more and more moves as temperature grows. The best way to improve these measurements therefore has been to increase the sample size, the number of Monte Carlo steps in each block, to have a better estimate of the physical quantities and damp the uncertainties.

We also compare the uncertainty of the estimate with temperature (the magnetic suceptibility is presented in another graph due to the difference of scale). We clearly see that fluctuations are more relevant at lower temperatures, especially when reguarding the spin-related quantities. Previous shorter simulations yielded, for $T^{\star} > 1$ ,an acceptable precision of the estimate, consistent between all the estimates; the increase of simulation length, on the other hand, heavily influenced the precision at low temperatures, which previously diverged.
